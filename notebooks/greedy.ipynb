```python
import matplotlib.pyplot as plt
import os
```

```python
from algos.greedy import *
```

```python
filename = "../data/d_pet_pictures.txt"
photos = load_data(filename)

slides = form_slides(photos)
del photos
print(f'Num slides : {len(slides)}')
slideshow = create_slideshow(slides, max_candidates=1000)
score = score_slideshow(slideshow)

print(f'Score : {score}')
```

```python
cum_scores = np.cumsum(slideshow.scores)

_, ax = plt.subplots(1, 1, figsize=(15, 5))
ax.plot(range(0, len(cum_scores)), cum_scores)
ax.set_title(f"Cumulative scores - {os.path.basename(filename)}")
ax.set_ylabel("Score")
ax.set_xlabel('Slides added');
```
